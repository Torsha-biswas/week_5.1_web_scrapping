```python
def scrap_func():
  # Check if the request was successful (status code 200)
  if response.status_code == 200:
      # Parse the HTML contnt of the page
      soup = BeautifulSoup(response.text, 'html.parser')

      # Extract book information
      books = []
      for book in soup.find_all('article', class_='product_pod'):
          title = book.h3.a.attrs['title']
          price = book.find('p', class_='price_color').text
          rating = book.p.attrs['class'][1]
          link = url + book.h3.a['href']
          books.append({'Title': title, 'Price': price, 'Rating': rating, 'Link': link})

      # Display the extracted data in a table
      table = PrettyTable()
      table.field_names = ["Title", "Price", "Rating", "Link"]

      for book in books:
          table.add_row([book['Title'], book['Price'], book['Rating'], book['Link']])

      print(table)
  else:
      print(f"Failed to retrieve the page. Status code: {response.status_code}")





















```

# BOOKS TO SCRAP PAGE 1

```python
# Send a GET request to the website
url = "http://books.toscrape.com/"
response = requests.get(url)
scrap_func()
```
